# Regular Expression Version

```python
def strToOneHot(vals):
    res = ""
    for val in vals:
        if val == "yes":
            res = res+ "1"
        else:
            res = res + "0"
    return (res)

def is_all_EN(vals):
    onehot = strToOneHot(vals)
    if(re.search('0',onehot)):
        return False
    else:
        return True

## 2nd Implementation - using a variable of the length of the array:
def is_all_EN(vals):
    onehot = strToOneHot(vals)
    l = str(len(vals))
    if(re.search('1{' + l + '}',onehot)):
        return True
    else:
        return False

## 3nd Implementation:
def is_all_EN(vals):
    onehot = strToOneHot(vals)
    if(re.search('^1+$',onehot)):
        return True
    else:
        return False

    
def is_all_nan(vals):
    onehot = strToOneHot(vals)
    if(re.search('1',onehot)):
        return False
    else:
        return True
    

def is_early_EN(vals):  ## checks that the vector starts with a sequence of 'yes's, followed by a seq of "nan" 
                        ## and less than half are "yes"
    onehot = strToOneHot(vals)   
    find = re.search('^1+0+$',onehot)     
    return ((find!=None) & (len(re.findall('1',onehot))<=len(onehot)/2))

def is_late_EN(vals):   
    onehot = strToOneHot(vals)
    find = re.search('^0+1+$',onehot)
    return (bool(find) & (len(re.findall('1',onehot))<=len(onehot)/2))

def is_intermit(vals):   #searches 1..10..01..1 or 0..01..10..0 substrings
    onehot = strToOneHot(vals)
    return bool(re.search('0+1+0+|1+0+1+',onehot))

def is_intermit(vals): #searches 101 or 010 substrings   
    onehot = strToOneHot(vals)
    return bool(re.search('010+|101',onehot))
    



def get_EN_pattern(vals):
    if is_all_EN(vals):
        return ("ALL_EN")
    elif is_early_EN(vals):
        return ("EARLY_EN")
    elif is_late_EN(vals):
        return ("LATE_EN")
    elif is_intermit(vals):
        return ("INTERMIT_EN")
    elif is_all_nan(vals):
        return ("ALL_NAN")
    else:
        return("OTHER")
    
    
def get_PN_pattern(vals):
    if is_all_EN(vals):
        return ("ALL_PN")
    elif is_early_EN(vals):
        return ("EARLY_PN")
    elif is_late_EN(vals):
        return ("LATE_PN")
    elif is_intermit(vals):
        return ("INTERMIT_PN")
    elif is_all_nan(vals):
        return ("ALL_NAN")
    else:
        return("OTHER")
    
    
def get_NG(vals):
    day = 0
    for val in vals:
        day = day+1
        if(float(val) >= 500):
            return (day)
    return(0)
    
```

### Adding new columns using the above functions### adding new columns using the above functions
#### There are two approaches presented here. The one that is in comment is suitable when the function receives several columns of the DF as input

```python
patient_level_df['EN'] = patient_level_df['ENs'].apply(get_EN_pattern) #patient_level_df.apply(lambda row: get_EN_pattern(row.ENs),axis=1)
patient_level_df['PN'] = patient_level_df['PNs'].apply(get_PN_pattern)  #patient_level_df.apply(lambda row: get_PN_pattern(row.PNs),axis=1)
patient_level_df['NG'] = patient_level_df['NGs'].apply(get_NG) #patient_level_df.apply(lambda row: get_NG(row.NGs),axis=1)
```
