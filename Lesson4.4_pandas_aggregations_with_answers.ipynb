### התא הבא כולל פונקציות שבאמצעותן נרצה להוסיף עמודות חדשות לטבלה החדשה על בסיס ערכי עמודות קיימים 
הקלטים לפונקציות הללו הם למעשה רשימות ערכים - כמו אלה שנוצרו בעמודות 
<br> ENs,PNs,NGs

```python
def strToOneHot(vals):
    res = ""
    for val in vals:
        if val == "yes":
            res = res+ "1"
        else:
            res = res + "0"
    return (res)

def is_all_EN(vals):
    for val in vals:
        if(val != "yes"):
            return (False)
    return(True)

def is_all_nan(vals):
    for val in vals:
        if(val == "yes"):
            return (False)
    return(True)

def is_early_EN(vals, n):    
    #check if early ENs
    isEarlyENs = True
    for i in range(n):
        if(vals[i]!='yes'):
            isEarlyENs = False
            break
    isRestNull = True
    for i in range(n,len(vals)):
        if (vals[i]!='nan'):
            isRestNull = False
    return (isEarlyENs & isRestNull)

    
def is_late_EN(vals, n):     
    #check if early ENs
    isEarlyNans = True
    for i in range(n):
        if(vals[i]!="nan"):
            isEarlyNans = False
            break
    isRestENs = True
    for i in range(n,len(vals)):
        if (vals[i]!='yes'):
            isRestENs = False
    return (isEarlyNans & isRestENs)

def is_intermit(vals):
    if len(vals)<=2:
        return False
    prev = vals[1]
    pre_prev = vals[0]
    for i in range(2,len(vals)):
        if ((pre_prev == "yes")&(prev=="nan")&(vals[i]=="yes"))|((pre_prev == "nan")&(prev=="yes")&(vals[i]=="nan")):
            return True
        else:
            pre_prev = prev
            prev = vals[i]
    return False


def is_intermit_gen(vals,n):
    if len(vals)<=n:
        return False
    for i in range(len(vals)-n):
        cur = vals[i]
        found=True
        for j in range(i+1,n+i):
            if cur == vals[j]:
                found = False
                break
            else:
                cur = vals[j]
        if found:
            return True

    return False


def get_EN_pattern(vals):
    if is_all_EN(vals):
        return ("ALL_EN")
    elif is_early_EN(vals):
        return ("EARLY_EN")
    elif is_late_EN(vals):
        return ("LATE_EN")
    elif is_intermit(vals):
        return ("INTERMIT_EN")
    elif is_all_nan(vals):
        return ("ALL_NAN")
    else:
        return("OTHER")
    
    
def get_PN_pattern(vals):
    if is_all_EN(vals):
        return ("ALL_PN")
    elif is_early_EN(vals):
        return ("EARLY_PN")
    elif is_late_EN(vals):
        return ("LATE_PN")
    elif is_intermit(vals):
        return ("INTERMIT_PN")
    elif is_all_nan(vals):
        return ("ALL_NAN")
    else:
        return("OTHER")
    
    
def get_NG(vals):
    day = 0
    for val in vals:
        day = day+1
        if(float(val) >= 500):
            return (day)
    return(0)
    
```

### Adding new columns using the above functions### adding new columns using the above functions
#### There are two approaches presented here. The one that is in comment is suitable when the function receives several columns of the DF as input

```python
patient_level_df['EN'] = patient_level_df['ENs'].apply(get_EN_pattern) #patient_level_df.apply(lambda row: get_EN_pattern(row.ENs),axis=1)
patient_level_df['PN'] = patient_level_df['PNs'].apply(get_PN_pattern)  #patient_level_df.apply(lambda row: get_PN_pattern(row.PNs),axis=1)
patient_level_df['NG'] = patient_level_df['NGs'].apply(get_NG) #patient_level_df.apply(lambda row: get_NG(row.NGs),axis=1)
```
